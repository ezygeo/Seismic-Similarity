```python
import numpy as np
import matplotlib.pyplot as plt
import similaritymeasures
```

```python
ti = 4
ns = 1501
t = np.linspace(0, ti, ns)
```

```python
f1 = 1
w1 = 2*np.pi*f1
y1 = np.sin(w1*t)
plt.subplot(141)
plt.plot(y1, t, 'r')
plt.ylim(4, 0)
plt.title('Trainer 1')
```

```python
f2 = 3
w2 = 2*np.pi*f2
y2 = np.sin(w2*t)
plt.subplot(142)
plt.plot(y2, t, 'g')
plt.ylim(4, 0)
plt.title('Trainer 2')
```

```python
f3 = 6
w3 = 2*np.pi*f3
y3 = np.sin(w3*t)
plt.subplot(143)
plt.plot(y3, t, 'b')
plt.ylim(4, 0)
plt.title('Trainer 3')
```

```python
f4 = 2
w4 = 2*np.pi*f4
y4 = np.sin(w4*t)
```

```python
xy1 = np.vstack((t, y1)).T
xy2 = np.vstack((t, y2)).T
xy3 = np.vstack((t, y3)).T
xy4 = np.vstack((t, y4)).T
```

```python
s1 = similaritymeasures.pcm(xy1, xy4)
s2 = similaritymeasures.pcm(xy2, xy4)
s3 = similaritymeasures.pcm(xy3, xy4)
```

```python
if s1 < s2 and s1 < s3:
    plt.subplot(144)
    plt.plot(y4, t, 'r')
    plt.ylim(4, 0)
    plt.title('Data')
elif s2 < s1 and s2 < s3:
    plt.subplot(144)
    plt.plot(y4, t, 'g')
    plt.ylim(4, 0)
    plt.title('Data')
elif s2 < s1 and s2 < s3:
    plt.subplot(144)
    plt.plot(y4, t, 'b')
    plt.ylim(4, 0)
    plt.title('Data')
```

### APPLY TO SEISMIC DATA

```python
data = np.genfromtxt('seismic.txt')
data = data.T
data = data[25:1526,:]
data = data[0:250]
```

```python
trainer1 = data[:,1]
trainer2 = data[:,30]
trainer3 = data[:,60]
np.shape(trainer1)
```

```python
t = 4*np.linspace(0, 249, 250)
```

```python
plt.subplot(141)
plt.plot(trainer1, t, 'r')
plt.ylim(1000, 0)
plt.title('Trainer 1')
```

```python
plt.subplot(142)
plt.plot(trainer2, t, 'g')
plt.ylim(1000, 0)
plt.title('Trainer 2')
```

```python
plt.subplot(143)
plt.plot(trainer1, t, 'b')
plt.ylim(1000, 0)
plt.title('Trainer 3')
```

```python
trainer1 = np.vstack((t, trainer1)).T
trainer2 = np.vstack((t, trainer2)).T
trainer3 = np.vstack((t, trainer3)).T
```

```python
for i in range(160):
    
    xy = np.vstack((t, data[:,i])).T
    s1 = similaritymeasures.pcm(trainer1, xy)
    s2 = similaritymeasures.pcm(trainer2, xy)
    s3 = similaritymeasures.pcm(trainer3, xy)  
    
    if s1 < s2 and s1 < s3:
        plt.plot(20*i + data[:,i], t, 'r')
        plt.ylim(1000, 0)
        plt.title('Data')
    elif s2 < s1 and s2 < s3:
        plt.plot(20*i + data[:,i], t, 'g')
        plt.ylim(1000, 0)
        plt.title('Data')
    elif s3 < s1 and s3 < s2:
        plt.plot(20*i + data[:,i], t, 'b')
        plt.ylim(1000, 0)
        plt.title('Data')
```
